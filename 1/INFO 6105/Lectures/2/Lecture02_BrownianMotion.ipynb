### Brownian Motion
The stochastic process $B(t)$ is a Brownian motion if 
- $B\left(0\right)=0$
- $B\left(t\right)-B\left(s\right)\sim N(0,\ \sigma^2(t-s))$ for any $t>s$
- If $t_1<t_2<t_3<t_4$ then $B\left(t_2\right)-B\left(t_1\right)$ is independent of $B\left(t_4\right)-B\left(t_3\right)$

```python
import numpy as np 
import matplotlib.pyplot as plt

T = 1000
N = 10
B = np.zeros((T,N))

for n in range(0,N):
    u = np.random.normal(0,1,T)
    B[:,n] = np.cumsum(u)

timevec = np.linspace(1,T,T)
plt.plot(timevec,B)
```
