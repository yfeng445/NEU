## Data Scaling

### 1 Z-Score Scaling (Standardization)
Z-score scaling (also known as standardization) transforms data to have a mean of 0 and a standard deviation of 1. This process is useful for normalizing data, especially when comparing features with different scales.

Here’s a Python script that demonstrates Z-score scaling:
- Generate Data: Create a sample dataset.
- Calculate Z-Scores: Compute the Z-scores for the data.
- Visualize: Plot the original and scaled data for comparison.

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the data sample
np.random.seed(0)  # For reproducibility
sample_size = 100  # Number of data points
mean_value = 50    # Mean of the data
std_dev = 15       # Standard deviation of the data

# Generate a random sample from a normal distribution
data = np.random.normal(loc=mean_value, scale=std_dev, size=sample_size)

# Calculate Z-scores
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Plot original and scaled data
plt.figure(figsize=(14, 6))

# Original data plot
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Scaled data plot (Z-scores)
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='lightgreen', edgecolor='black')
plt.title('Z-Score Scaled Data')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
````

### 2 Min-Max Scaling
Min-max scaling (also known as normalization) rescales data to fit within a specified range, typically $[0, 1]$. This process is useful for normalizing features, especially when features have different units or scales.

Here’s a Python script that demonstrates min-max scaling:
- Generate Data: Create a sample dataset.
- Apply Min-Max Scaling: Transform the data to a specified range.
- Visualize: Plot the original and scaled data for comparison.

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the data sample
np.random.seed(0)  # For reproducibility
sample_size = 100  # Number of data points
mean_value = 50    # Mean of the data
std_dev = 15       # Standard deviation of the data

# Generate a random sample from a normal distribution
data = np.random.normal(loc=mean_value, scale=std_dev, size=sample_size)

# Min-Max Scaling
min_value = np.min(data)
max_value = np.max(data)
scaled_data = (data - min_value) / (max_value - min_value)

# Plot original and scaled data
plt.figure(figsize=(14, 6))

# Original data plot
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Scaled data plot (Min-Max Scaling)
plt.subplot(1, 2, 2)
plt.hist(scaled_data, bins=30, color='lightgreen', edgecolor='black')
plt.title('Min-Max Scaled Data')
plt.xlabel('Scaled Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
````

**Explanation:**
- Generate Data:```np.random.normal(loc=mean_value, scale=std_dev, size=sample_size)``` generates a dataset with a specified mean and standard deviation.
- Apply Min-Max Scaling: 
    - Minimum and Maximum Values: Compute the minimum and maximum values of the original data.
    - Scaling Formula: For each data point, apply the formula: $scaled_{data} =\frac{data − min}{max − min}$. This scales the data to the range [0, 1].
- Plotting:
    - Original Data: A histogram of the original data to show its distribution.
    - Scaled Data: A histogram of the scaled data to show how it fits within the [0, 1] range after min-max scaling.
This script will help you visualize the effect of min-max scaling on the dataset. Ensure you have numpy and matplotlib installed (pip install numpy matplotlib) to run this code successfully.

### 3 Log Transformation
Log transformation is a technique used to stabilize the variance and make data more normally distributed. It is particularly useful when dealing with positively skewed data.

Here’s a Python script that demonstrates log transformation:
- Generate Data: Create a sample dataset with skewness.
- Apply Log Transformation: Transform the data using the logarithm function.
- Visualize: Plot the original and transformed data for comparison.

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the data sample
np.random.seed(0)  # For reproducibility
sample_size = 100  # Number of data points

# Generate a skewed sample of data using an exponential distribution
data = np.random.exponential(scale=1.0, size=sample_size)*100000

# Apply Log Transformation
# Adding a small constant (1e-9) to avoid log(0) issues
log_transformed_data = np.log(data + 1e-9)

# Plot original and transformed data
plt.figure(figsize=(14, 6))

# Original data plot
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Log-transformed data plot
plt.subplot(1, 2, 2)
plt.hist(log_transformed_data, bins=30, color='lightgreen', edgecolor='black')
plt.title('Log Transformed Data')
plt.xlabel('Log(Value)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
````

### 4 Square Root Transformation
A square root transformation is used to stabilize variance and reduce skewness in data. It is particularly useful for count data or data with positive skew.

Here’s a Python script that demonstrates how to apply a square root transformation:
- Generate Data: Create a sample dataset, potentially with skewness.
- Apply Square Root Transformation: Transform the data using the square root function.
- Visualize: Plot the original and transformed data for comparison.

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the data sample
np.random.seed(0)  # For reproducibility
sample_size = 100  # Number of data points

# Generate a skewed sample of data using an exponential distribution
data = np.random.exponential(scale=1.0, size=sample_size)

# Apply Square Root Transformation
# Adding a small constant (1e-9) to avoid sqrt(0) issues
sqrt_transformed_data = np.sqrt(data + 1e-9)

# Plot original and transformed data
plt.figure(figsize=(14, 6))

# Original data plot
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data with Skewness')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Square root transformed data plot
plt.subplot(1, 2, 2)
plt.hist(sqrt_transformed_data, bins=30, color='lightgreen', edgecolor='black')
plt.title('Square Root Transformed Data')
plt.xlabel('Square Root(Value)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
````

**Explanation:**
- Generate Data: np.random.exponential(scale=1.0, size=sample_size) generates a dataset from an exponential distribution, which is skewed.
- Apply Square Root Transformation:
    - To handle potential issues with zero values (since the square root of zero is zero), a small constant (1e-9) is added.
    - np.sqrt(data + 1e-9) applies the square root transformation to each data point.
- Plotting:
    - Original Data: A histogram of the original skewed data to show its distribution.
    - Square Root Transformed Data: A histogram of the transformed data to show how the square root transformation has adjusted the distribution, making it less skewed.

This script provides a visual comparison of the data before and after applying the square root transformation, helping to illustrate the impact of the transformation on skewness and variance. Ensure you have numpy and matplotlib installed (pip install numpy matplotlib) to run this code successfully.

### 5 Robust Scaling
Robust scaling is a technique used to scale features by removing the median and scaling according to the Interquartile Range (IQR). This method is less sensitive to outliers compared to standard scaling methods that rely on the mean and standard deviation.

Here's a Python script to demonstrate robust scaling:
- Generate Data: Create a sample dataset with outliers.
- Apply Robust Scaling: Transform the data using robust scaling.
- Visualize: Plot the original and scaled data for comparison.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler

# Parameters for the data sample
np.random.seed(0)  # For reproducibility
sample_size = 100  # Number of data points

# Generate a random sample from a normal distribution and add outliers
data = np.concatenate([np.random.normal(loc=0, scale=1, size=sample_size - 10), 
                        np.random.normal(loc=20, scale=5, size=10)])

# Apply Robust Scaling
scaler = RobustScaler()
data_reshaped = data.reshape(-1, 1)  # Reshape data for scaler
scaled_data = scaler.fit_transform(data_reshaped).flatten()  # Flatten back to original shape

# Plot original and scaled data
plt.figure(figsize=(14, 6))

# Original data plot
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data with Outliers')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Scaled data plot (Robust Scaling)
plt.subplot(1, 2, 2)
plt.hist(scaled_data, bins=30, color='lightgreen', edgecolor='black')
plt.title('Robust Scaled Data')
plt.xlabel('Robust Scaled Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
````

**Explanation:**
- Generate Data:
    - ```np.random.normal(loc=0, scale=1, size=sample_size - 10)``` generates the bulk of the data.
    - ```np.random.normal(loc=20, scale=5, size=10)``` generates outliers with a different mean and standard deviation.
    - ```np.concatenate``` combines the two parts to form the complete dataset.
- Apply Robust Scaling:
    - ```RobustScaler from sklearn.preprocessing``` scales the data based on the median and the IQR.
    - the scaled value is calculated using the median and the Interquartile Range (IQR) of the data. The formula for robust scaling is: <br> $scaled_{data} = \frac{data − median(data)}{IQR}$ <br>
    - $IQR = Q3 - Q1$ (Interquartile Range) is the difference between the 75th percentile ($Q3$) and the 25th percentile ($Q1$) of the data.
    - $Q1$ (25th percentile) is the value below which 25% of the data falls. $Q3$ (75th percentile) is the value below which 75% of the data falls. Median is the middle value of the data when sorted.
    - Data needs to be reshaped to 2D for the scaler (data.reshape(-1, 1)), and then it's transformed and flattened back to 1D.
- Plotting:
    - Original Data: A histogram showing the original data distribution with outliers.
    - Scaled Data: A histogram showing the distribution of the robustly scaled data. The scaling should make the data less sensitive to outliers, and the histogram should reflect this normalization.

This script demonstrates how robust scaling can adjust the data, particularly in the presence of outliers, making it more suitable for further analysis. Ensure you have numpy, matplotlib, and scikit-learn installed (```pip install numpy matplotlib scikit-learn```) to run this code successfully.
