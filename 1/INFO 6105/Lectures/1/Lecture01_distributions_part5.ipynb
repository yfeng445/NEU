### 1 Exponential Distribution

**Explanation:**

PDF (Probability Density Function): The formula for the PDF of an exponential distribution is given by:

$f(x; \lambda) = \lambda e^{-\lambda x}$ 

where lambda is the rate parameter. The larger lambda is, the faster the distribution decays.

CDF (Cumulative Distribution Function): The CDF shows the probability that a random variable from the distribution is less than or equal to x. It is given by: 

$F(x; \lambda) = 1 - e^{-\lambda x}$, $x \geq 0$.  

Mean and Variance:
- Mean: $\mu = \frac{1}{\lambda}$
- Variance: $\sigma^{2} = \frac{1}{\lambda^{2}}$

The exponential distribution is memoryless, meaning that the probability of an event occurring in the next time period is independent of how much time has already passed.

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the rate parameter (lambda) for the exponential distribution
rate = 1.5  # Lambda (rate) controls the distribution

# Generate x values for the plot
x = np.linspace(0, 4, 1000)

# PDF of exponential distribution
pdf = rate * np.exp(-rate * x)

# CDF of exponential distribution
cdf = 1 - np.exp(-rate * x)

# Theoretical mean and variance
mean = 1 / rate
variance = 1 / (rate ** 2)

# Plot the PDF and CDF
plt.figure(figsize=(12, 6))

# Plot PDF
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label=f"PDF (lambda={rate})", color='b', linewidth = 4)
plt.title("Exponential Distribution - PDF")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label=f"CDF (lambda={rate})", color='g', linewidth = 4)
plt.title("Exponential Distribution - CDF")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.grid(True)

plt.tight_layout()
plt.show()

# Display properties
print(f"Exponential Distribution Properties:")
print(f"Rate (\u03BB): {rate}")
print(f"Mean (1/\u03BB): {mean}")
print(f"Variance (1/\u03BB\u00B2): {variance}")
````

#### Understanding $\lambda$

In the context of an exponential distribution, the rate parameter $\lambda$ plays a crucial role in determining the shape and behavior of the distribution. Here's how to understand
$\lambda$ with a few practical examples:

**a. Waiting Time for a Bus:** Imagine you're at a bus stop where buses arrive on average every 10 minutes. The exponential distribution can model the waiting time between bus arrivals.
- If the average waiting time is 10 minutes, the rate $\lambda$ is the reciprocal of the average time, so 

$\lambda = \frac{1}{10} = 0.1$ (per minute).

- A higher $\lambda$ means buses arrive more frequently (shorter waiting time). A lower $\lambda$ means buses are less frequent.

**b. Decay of Radioactive Particles:** The exponential distribution can also describe the time between radioactive decays of particles in a sample.
- Suppose the average decay time of a radioactive particle is 5 hours, then 

    $\lambda = \frac{1}{5} = 0.2$ (decays per hour).

- If $\lambda$ is small, the decays happen less frequently (particles decay slowly). If $\lambda$ is large, the decays happen more frequently (particles decay faster).

**c. Lifespan of Electronic Devices:** The time until an electronic device (such as a light bulb or a machine component) fails can be modeled using an exponential distribution.
- If a light bulb lasts an average of $2000$ hours, then 

    $\lambda = \frac{1}{2000} = 0.0005$ (failures per hour).

A lower $\lambda$ indicates a longer lifespan (fewer failures per hour). A higher $\lambda$ means the device fails more frequently (shorter lifespan).

**d. Customer Service: Time Between Calls:**
In a call center, the time between consecutive customer calls can be modeled using an exponential distribution.
- If the call center receives 30 calls per hour on average, 

    $\lambda = 30$ (calls per hour).

A higher $\lambda$ means the center receives calls more frequently, and the time between calls is shorter. A lower $\lambda$ means calls are less frequent, with longer gaps between them.

In all these examples, $\lambda$ controls how often an event happens in a given time period or how fast something occurs (e.g., decay, failure, arrival, etc.).

#### Exponential distributions with three different values of $\lambda$ (rate parameter) 

````python
import numpy as np
import matplotlib.pyplot as plt

# Set three different lambda values (rate parameters)
lambdas = [0.5, 1.0, 1.5]

# Generate a range of x values
x = np.linspace(0, 10, 1000)

# Plot the exponential distributions for each lambda
plt.figure(figsize=(10, 6))

for lam in lambdas:
    pdf = lam * np.exp(-lam * x)  # PDF of the exponential distribution
    plt.plot(x, pdf, label=f' = {lam}', linewidth = 4)

# Customize the plot
plt.title('Exponential Distributions with Different  Values')
plt.xlabel('x')
plt.ylabel('Probability Density Function (PDF)')
plt.axis([0, 10, 0, 1.5])
plt.legend()
plt.grid(True)
plt.show()
````

### 2 Gamma Distribution

Here is Python code to demonstrate the properties of the Gamma distribution, including the Probability Density Function (PDF) for different shape (
$$) and scale ($\theta$) parameters, along with its mean and variance.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Define shape and scale parameters for the Gamma distribution
parameters = [
    {"shape": 2, "scale": 1, "label": "k=2, =1"},
    {"shape": 5, "scale": 1, "label": "k=5, =1"},
    {"shape": 9, "scale": 0.5, "label": "k=9, =0.5"}
]

# Generate a range of x values
x = np.linspace(0, 20, 1000)

# Plot the Gamma distributions for different shape and scale parameters
plt.figure(figsize=(10, 6))

for param in parameters:
    k, theta = param["shape"], param["scale"]
    pdf = gamma.pdf(x, k, scale=theta)  # PDF of the Gamma distribution
    plt.plot(x, pdf, label=param["label"], linewidth = 4)

# Customize the plot
plt.title('Gamma Distributions with Different Shape and Scale Parameters')
plt.xlabel('x')
plt.ylabel('Probability Density Function (PDF)')
plt.axis([0, 20, 0, 0.4])
plt.legend()
plt.grid(True)
plt.show()

# Calculate and display properties (mean and variance)
for param in parameters:
    k, theta = param["shape"], param["scale"]
    mean = k * theta  # Mean of the Gamma distribution
    variance = k * theta ** 2  # Variance of the Gamma distribution
    print(f"For k={k}, \u03B8 ={theta}: Mean = {mean}, Variance = {variance}")
````

**Explanation:**
- Parameters $k$ and $\theta$: 
    - $\theta$: The Gamma distribution is characterized by its shape parameter.  
    - $k$ the scale parameter.  
Here, we visualize Gamma distributions for three different combinations of $k$ and $\theta$. 

- PDF: The Probability Density Function (PDF) is computed using ```scipy.stats.gamma.pdf()```. This generates the distribution curve based on the shape and scale parameters.
- Mean and Variance:
    - Mean: $\mu = k\times \theta$
    - Variance: $\sigma^{2}= k\times \theta^{2}$
 
This code will plot the Gamma distribution for the given parameter combinations, and it will also print out the mean and variance for each distribution. The shape of the Gamma distribution changes significantly with different values of $k$ and $\theta$.
