### 1 Chi-square Distribution

Here’s Python code to visualize the properties of the Chi-square distribution, such as its Probability Density Function (PDF) for different degrees of freedom, and its mean and variance.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Define degrees of freedom for the Chi-square distribution
degrees_of_freedom = [2, 5, 10]

# Generate a range of x values
x = np.linspace(0, 20, 1000)

# Plot the Chi-square distributions for different degrees of freedom
plt.figure(figsize=(10, 6))

for df in degrees_of_freedom:
    pdf = chi2.pdf(x, df)  # PDF of the Chi-square distribution
    plt.plot(x, pdf, label=f'df = {df}', linewidth = 4)

# Customize the plot
plt.title('Chi-square Distributions with Different Degrees of Freedom')
plt.axis([0, 20, 0, 0.5])
plt.xlabel('x')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()
plt.grid(True)
plt.show()

# Calculate and display properties
for df in degrees_of_freedom:
    mean = df  # Mean of the Chi-square distribution is equal to the degrees of freedom
    variance = 2 * df  # Variance is 2 times the degrees of freedom
    print(f"For df = {df}: Mean = {mean}, Variance = {variance}")
````

**Explanation:**
- Degrees of Freedom (df): The Chi-square distribution is parameterized by its degrees of freedom. The larger the degrees of freedom, the more the distribution shifts to the right.
- PDF: The code uses ```scipy.stats.chi2.pdf()``` to compute the Probability Density Function (PDF) of the Chi-square distribution for different degrees of freedom.
- Mean and Variance: The mean of the Chi-square distribution is equal to the degrees of freedom (mean=df), and the variance is twice the degrees of freedom (variance= 2 × df).

The plot will show how the shape of the Chi-square distribution changes for degrees of freedom (df = 2, 5, 10). A Chi-square distribution with smaller df is skewed to the right, while larger df makes it more symmetric.

### 2 Weibull Distribution

Here’s Python code to demonstrate the properties of the Weibull distribution, including its Probability Density Function (PDF) for different shape ($c$) and scale ($\lambda$) parameters, along with its mean and variance:

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min
from scipy.special import gamma

# Define shape and scale parameters for the Weibull distribution
parameters = [
    {"shape": 1.5, "scale": 1, "label": "c=1.5, λ=1"},
    {"shape": 2.0, "scale": 1, "label": "c=2.0, λ=1"},
    {"shape": 3.0, "scale": 1.5, "label": "c=3.0, λ=1.5"}
]

# Generate a range of x values
x = np.linspace(0, 5, 1000)

# Plot the Weibull distributions for different shape and scale parameters
plt.figure(figsize=(10, 6))

for param in parameters:
    c, lam = param["shape"], param["scale"]
    pdf = weibull_min.pdf(x, c, scale=lam)  # PDF of the Weibull distribution
    plt.plot(x, pdf, label=param["label"], linewidth = 4)

# Customize the plot
plt.title('Weibull Distributions with Different Shape and Scale Parameters')
plt.xlabel('x')
plt.ylabel('Probability Density Function (PDF)')
plt.axis([0, 5, 0, 1])
plt.legend()
plt.grid(True)
plt.show()

# Calculate and display properties (mean and variance)
for param in parameters:
    c, lam = param["shape"], param["scale"]
    mean = lam * gamma(1 + 1/c)  # Mean of the Weibull distribution
    variance = lam**2 * (gamma(1 + 2/c) - (gamma(1 + 1/c))**2)  # Variance of the Weibull distribution
    print(f"For c={c}, λ={lam}: Mean = {mean:.2f}, Variance = {variance:.2f}")
````

**Explanation:**
- Parameters $c$ and $\lambda$: The Weibull distribution is characterized by its shape parameter $𝑐$ and scale parameter $\lambda$. We visualize Weibull distributions for different combinations of $c$ and $\lambda$. 

- PDF: The Probability Density Function (PDF) is computed using ```scipy.stats.weibull_min.pdf()```. This generates the distribution curve based on the shape and scale parameters.

- Mean and Variance:
    - Mean: $\mu = \lambda\Gamma(1+\frac{1}{c})$, where $\Gamma$ is the Gamma function. 
    - Variance: $\sigma^{2} = \lambda^{2}\big[\Gamma(1+\frac{2}{c})-(\Gamma(1+\frac{1}{c}))^{2} \big]$

This code will plot the Weibull distribution for the given parameter combinations and print out the mean and variance for each distribution. The shape of the Weibull distribution varies with different values of $c$ and $\lambda$, affecting its skewness and scale.<br>

### 3 Beta Distribution

Here’s Python code to demonstrate the properties of the Beta distribution, including its Probability Density Function (PDF) for different shape parameters ($\alpha$ and $\beta$), along with its mean and variance:

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Define shape parameters for the Beta distribution
parameters = [
    {"alpha": 2, "beta": 2, "label": "α=2, β=2"},
    {"alpha": 5, "beta": 1, "label": "α=5, β=1"},
    {"alpha": 1, "beta": 5, "label": "α=1, β=5"}
]

# Generate a range of x values
x = np.linspace(0, 1, 1000)

# Plot the Beta distributions for different shape parameters
plt.figure(figsize=(10, 6))

for param in parameters:
    a, b = param["alpha"], param["beta"]
    pdf = beta.pdf(x, a, b)  # PDF of the Beta distribution
    plt.plot(x, pdf, label=param["label"], linewidth = 4)

# Customize the plot
plt.title('Beta Distributions with Different Shape Parameters')
plt.axis([0, 1, 0, 5])
plt.xlabel('x')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()
plt.grid(True)
plt.show()

# Calculate and display properties (mean and variance)
for param in parameters:
    a, b = param["alpha"], param["beta"]
    mean = a / (a + b)  # Mean of the Beta distribution
    variance = (a * b) / ((a + b)**2 * (a + b + 1))  # Variance of the Beta distribution
    print(f"For α={a}, β={b}: Mean = {mean:.2f}, Variance = {variance:.2f}")
````

**Explanation:**
- Parameters $\alpha$ and $\beta$: The Beta distribution is characterized by its shape parameters $\alpha$ and $\beta$. We visualize Beta distributions for different combinations of $\alpha$ and $\beta$. 

- PDF: The Probability Density Function (PDF) is computed using ```scipy.stats.beta.pdf()```. This generates the distribution curve based on the shape parameters.

- Mean and Variance:
    - Mean: $\mu = \frac{\alpha}{\alpha + \beta}$ 
    - Variance: $\sigma^{2} = \frac{\alpha\beta}{(\alpha+\beta)^{2}(\alpha+\beta+1)}$
    
This code will plot the Beta distribution for the specified parameter combinations and print out the mean and variance for each distribution. The shape of the Beta distribution changes with different values of $\alpha$ and 
$\beta$, affecting its skewness and peak.
