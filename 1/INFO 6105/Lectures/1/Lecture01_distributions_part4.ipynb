### 3 Log-Normal Distribution

- Definition: A random variable $X$ is said to have a lognormal distribution if its logarithm $Y= \ln‚Å°{(X)}$ is normally distributed. 
- Parameters: The log-normal distribution is characterized by two parameters: the mean ($\mu$) and standard deviation ($\sigma$) of the underlying normal distribution.
- Probability Density Function (PDF): Shows the shape of the log-normal distribution.
- Cumulative Distribution Function (CDF): Shows the cumulative probability up to a given value.
- Sample Data: Generates and visualizes a sample from the log-normal distribution.
- Histogram: Displays the distribution of sampled data.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Parameters for the log-normal distribution
mu = 0      # Mean of the underlying normal distribution
sigma = 0.954  # Standard deviation of the underlying normal distribution (œÉ = 1 for ln(Œº))

# Define the log-normal distribution
dist = lognorm(sigma, scale=np.exp(mu))

# Define the range for plotting
x = np.linspace(0, 5, 1000)

# Compute PDF and CDF
pdf = dist.pdf(x)
cdf = dist.cdf(x)

# Generate sample data from the log-normal distribution
sample_size = 1000
samples = dist.rvs(size=sample_size)

# Plot the PDF
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
plt.plot(x, pdf, 'b-', lw=2)
plt.title('Probability Density Function (PDF)')
plt.xlabel('x')
plt.ylabel('f(x)')

# Plot the CDF
plt.subplot(2, 2, 2)
plt.plot(x, cdf, 'r-', lw=2)
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('x')
plt.ylabel('F(x)')

# Plot histogram of the samples
plt.subplot(2, 2, 3)
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title('Histogram of Sampled Data')
plt.xlabel('x')
plt.ylabel('Frequency')

# Plot log-normal distribution area
plt.subplot(2, 2, 4)
plt.fill_between(x, 0, pdf, color='skyblue', alpha=0.4)
plt.title('Log-Normal Distribution Area')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.tight_layout()
plt.show()
```

**Explanation:**
- Parameters:
    - mu is the mean of the underlying normal distribution, which is the logarithm of the log-normal distribution.
    - sigma is the standard deviation of the underlying normal distribution.
- Distributions:
    - lognorm(sigma, scale=np.exp(mu)) creates a log-normal distribution with the specified parameters. The scale parameter is exp(mu), which adjusts for the mean of the log-transformed normal distribution.
- PDF and CDF:
    - pdf is the Probability Density Function of the log-normal distribution.
    - cdf is the Cumulative Distribution Function of the log-normal distribution.
- Sample Data: Generate 1000 samples from the log-normal distribution.
- Plots:
    - PDF Plot: Shows the shape of the log-normal distribution curve.
    - CDF Plot: Shows the cumulative probability up to each x value.
    - Histogram: Displays the distribution of the sampled data.
    - Fill Between Plot: Highlights the area under the PDF curve.

### 4 Student's t Distribution

- Definition and Parameters: Degrees of freedom (df).
- Probability Density Function (PDF): The shape of the distribution.
- Cumulative Distribution Function (CDF): The cumulative probability up to a given value.
- Sample Data: Generates and visualizes a sample from the t-distribution.
- Histogram: Displays the distribution of sampled data.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Parameters for the Student's t-distribution
df = 10  # Degrees of freedom

# Define the t-distribution
dist = t(df)

# Define the range for plotting
x = np.linspace(-5, 5, 1000)

# Compute PDF and CDF
pdf = dist.pdf(x)
cdf = dist.cdf(x)

# Generate sample data from the t-distribution
sample_size = 1000
samples = dist.rvs(size=sample_size)

# Plot the PDF
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
plt.plot(x, pdf, 'b-', lw=2)
plt.title('Probability Density Function (PDF)')
plt.xlabel('x')
plt.ylabel('f(x)')

# Plot the CDF
plt.subplot(2, 2, 2)
plt.plot(x, cdf, 'r-', lw=2)
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('x')
plt.ylabel('F(x)')

# Plot histogram of the samples
plt.subplot(2, 2, 3)
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title('Histogram of Sampled Data')
plt.xlabel('x')
plt.ylabel('Frequency')

# Plot t-distribution area
plt.subplot(2, 2, 4)
plt.fill_between(x, 0, pdf, color='skyblue', alpha=0.4)
plt.title('Student\'s t-Distribution Area')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.tight_layout()
plt.show()
```

**Explanation:**
- Parameters: 
     - df represents the degrees of freedom, which controls the shape of the t-distribution. Lower df values result in a distribution with heavier tails.
- PDF and CDF:
     - pdf is the Probability Density Function of the t-distribution. 
     - cdf is the Cumulative Distribution Function of the t-distribution.
- Sample Data:
    - We generate 1000 samples from the t-distribution using dist.rvs(size=sample_size).
- Plots:
    - PDF Plot: Shows the shape of the t-distribution curve.
    - CDF Plot: Shows the cumulative probability up to each x value.
    - Histogram: Displays the distribution of the sampled data.
    - Fill Between Plot: Highlights the area under the PDF curve, similar to how the area under a probability density function can be visualized.

This script will provide a comprehensive visualization of the Student's t-distribution and its properties. Make sure you have the necessary libraries (numpy, matplotlib, and scipy) installed in your Python environment to run this code (pip install numpy matplotlib scipy).

### 5 Student's t Distribution vs Normal Distribution
The exmaple below showing the following:
- Comparison of PDFs: Compare the PDFs of the t-distribution with different degrees of freedom to the normal distribution.
- Convergence: Show how the t-distribution approaches the normal distribution as the degrees of freedom increase.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# Parameters for the normal and t-distributions
df_values = [1, 5, 10, 30]  # Different degrees of freedom for t-distribution
mu = 0  # Mean for the normal distribution
sigma = 1  # Standard deviation for the normal distribution

# Define the normal distribution
normal_dist = norm(loc=mu, scale=sigma)

# Define the range for plotting
x = np.linspace(-5, 5, 1000)

# Create a plot
plt.figure(figsize=(14, 8))

# Plot PDFs of t-distributions with different degrees of freedom
for df in df_values:
    t_dist = t(df)
    pdf_t = t_dist.pdf(x)
    plt.plot(x, pdf_t, label=f"t-distribution df={df}")

# Plot PDF of the normal distribution
pdf_normal = normal_dist.pdf(x)
plt.plot(x, pdf_normal, 'k--', label='Normal distribution', lw=2)

# Add labels and title
plt.title('Comparison of Student\'s t-Distribution with Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

**Explanation:**
- Parameters: df_values represents different degrees of freedom for the t-distribution. These values are used to show how the shape of the t-distribution changes with varying df.
- Distributions: normal_dist is the standard normal distribution with mean $\mu$ and standard deviation $\sigma$.For each df in df_values, a t-distribution is created and its PDF is computed.
- Plotting: The script plots the PDFs of t-distributions with different degrees of freedom along with the PDF of the normal distribution (k-- for dashed line). As df increases, the shape of the t-distribution becomes closer to the normal distribution.

**Observations:**
- Heavy Tails: For lower degrees of freedom (e.g., df=1), the t-distribution has heavier tails compared to the normal distribution. This reflects greater uncertainty or variability in estimates when sample sizes are small.
- Convergence: As the degrees of freedom increase (e.g., df=30), the t-distribution approaches the shape of the normal distribution, reflecting the fact that with a larger sample size, the sample mean distribution approximates a normal distribution.

This script provides a clear visual comparison of how the Student's t-distribution with varying degrees of freedom relates to the normal distribution, illustrating the convergence of the t-distribution to the normal distribution as df increases.
